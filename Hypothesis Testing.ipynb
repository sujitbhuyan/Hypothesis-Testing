### Pearson Correlation: Temperature vs. Bike Count
Hypothesis:

Null Hypothesis (H0): There is no significant correlation between temperature and bike count.

Alternative Hypothesis (H1): There is a significant correlation between temperature and bike count.

```python
from scipy.stats import pearsonr

# Extracting the target variable and temperature
temperature = df['temperature']
count = df['count']

# Performing Pearson's correlation test
corr_coeff, p_value_corr = pearsonr(temperature, count)

print("Pearson Correlation Test:")
print(f"Pearson correlation coefficient: {corr_coeff}")
print(f"P-value: {p_value_corr}")
```

### Spearman Correlation: Temperature vs. Bike Count
    
Hypothesis:

Null Hypothesis (H0): There is no significant correlation between temperature and bike count.

Alternative Hypothesis (H1): There is a significant correlation between temperature and bike count.

```python
from scipy.stats import spearmanr

# Extracting the target variable and temperature
count = df['count']
temperature = df['temperature']

# Performing Spearman's correlation test
corr_coeff, p_value_corr = spearmanr(temperature, count)

print("Spearman Correlation Test:")
print(f"Spearman correlation coefficient: {corr_coeff}")
print(f"P-value: {p_value_corr}")
```

If the p-value is less than the significance level (e.g., 0.05), we can conclude that there is a significant correlation between temperature and  count.

### ANOVA: Comparing Mean Bike Count Across Different Seasons
    
Hypothesis:

Null Hypothesis (H0): There is no significant difference in the mean bike count across different seasons.

Alternative Hypothesis (H1): There is a significant difference in the mean bike count across different seasons.

```python
from scipy.stats import f_oneway

# Extracting the target variable and season status
count = df['count']
season = df['season']

# Splitting the data into groups based on seasons
count_season_1 = count[season == 1]
count_season_2 = count[season == 2]
count_season_3 = count[season == 3]
count_season_4 = count[season == 4]

# Performing a one-way ANOVA
f_stat, p_value_anova = f_oneway(count_season_1, count_season_2, count_season_3, count_season_4)

print("ANOVA Test:")
print(f"F-statistic: {f_stat}")
print(f"P-value: {p_value_anova}")
```

If the p-value is less than the significance level (e.g., 0.05), we can reject the null hypothesis and conclude that there is a significant difference in the mean count across different seasons.

### T-test for Difference in Means: Holidays vs. Non-Holidays
    
Hypothesis:

Null Hypothesis (H0): There is no significant difference in the mean bike count between holidays and non-holidays.

Alternative Hypothesis (H1): There is a significant difference in the mean bike count between holidays and non-holidays.

```python
from scipy.stats import ttest_ind

# Extracting the target variable and holiday status
count = df['count']
holiday = df['holiday']

# Splitting the data into two groups: holiday and non-holiday
count_holiday = count[holiday == 1]
count_non_holiday = count[holiday == 0]

# Performing an independent samples t-test
t_stat, p_value = ttest_ind(count_holiday, count_non_holiday)

print("T-test for Difference in Means:")
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")
```

If the p-value is less than the significance level (e.g., 0.05), we can reject the null hypothesis and conclude that there is a significant difference in the count between holidays and non-holidays
